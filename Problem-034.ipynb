Problem Statement - 34 | Find the First and Last Position of Sorted Array

```python
class Solution {
    public int[] searchRange(int[] nums, int target) {
        return new int[]{find(nums, target, true), find(nums, target, false)};
    }
    int find(int[] nums, int target, boolean first) {
        int s = 0, e = nums.length - 1, ans = -1;
        while (s <= e) {
            int m = (s + e) / 2;

            if (nums[m] == target) {
                ans = m;
                if (first) e = m - 1;
                else s = m + 1;
            } else if (nums[m] < target)
                s = m + 1;
            else
                e = m - 1;
        }
        return ans;
    }
}
```
